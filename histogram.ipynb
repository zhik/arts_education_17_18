```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ARTS_FILE = './2017-2018 Annual Arts In Schools Report (Cultural & Arts Partners with highlights) 9.11.19.xlsx'
xl = pd.ExcelFile(ARTS_FILE)
sheet_names = xl.sheet_names

all_arts = None
for name in sheet_names:
    sheet = pd.read_excel(ARTS_FILE, name)
    if all_arts is None:
        #init
        all_arts = sheet
    else:
        #append
        all_arts = all_arts.append(sheet, ignore_index=True, sort=False)
all_arts.tail()

#drop empty rows
all_arts = all_arts.dropna(subset=['DBN'], axis = 0).reset_index(drop=True).rename(columns=lambda x: x.strip())
```
