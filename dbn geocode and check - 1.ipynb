```python
import pandas as pd
import requests
```

```python
#read all sheets and get unqiue DBNs
ARTS_FILE = './2017-2018 Annual Arts In Schools Report.xlsx'
xl = pd.ExcelFile(ARTS_FILE)
sheet_names = xl.sheet_names

all_dbns = set()
for name in sheet_names:
    sheet = pd.read_excel(ARTS_FILE, name)
    dbns = set(str(dbn).strip().upper() for dbn in sheet['DBN'])
    all_dbns = all_dbns|dbns
```

```python
school_locations = pd.read_csv('./2017_-_2018_School_Locations.csv')
school_locations.head()
```

```python
#check if there are any missing DBNs, by doing a merge
all_dbns_df = pd.DataFrame(all_dbns, columns=['DBN'])

#change type and clean up "ATS SYSTEM CODE"
school_locations['DBN'] = school_locations.apply(lambda row: str(row["ATS SYSTEM CODE"]).strip().upper(), axis =1)

merge = pd.merge(school_locations, all_dbns_df, on="DBN", how="right")
merge.head()
```
