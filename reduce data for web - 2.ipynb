```python
import pandas as pd
import re
```

```python
ARTS_FILE = './2017-2018 Annual Arts In Schools Report.xlsx'
xl = pd.ExcelFile(ARTS_FILE)
sheet_names = xl.sheet_names

all_arts = None
for name in sheet_names:
    sheet = pd.read_excel(ARTS_FILE, name)
    if all_arts is None:
        #init
        all_arts = sheet
    else:
        #append
        all_arts = all_arts.append(sheet, ignore_index=True, sort=False)
all_arts.tail()
```

```python
#drop empty rows
all_arts = all_arts.dropna(subset=['DBN'], axis = 0).reset_index(drop=True).rename(columns=lambda x: x.strip())
```

```python
#strip all text cells extra space
all_arts = all_arts.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
all_arts.to_csv('./processed/all_arts.csv', index= False)
```

```python
school_locations = pd.read_csv('./2017_-_2018_School_Locations.csv')
school_locations.head()
```

```python
school_locations['DBN'] = school_locations.apply(lambda row: str(row["ATS SYSTEM CODE"]).strip().upper(), axis =1)

def getLocation(row):
    string = str(row['Location 1'])
    try:
        location = re.search("\(([^)]+)\)", string).group(1)
        return pd.Series([float(coord) for coord in location.split(',')])
    except:
        print('error, {} - with value {}'.format(row['DBN'], string))
    
    
school_locations[["lat", "lng"]] = school_locations.apply(getLocation, axis = 1)

mn_school_locations = school_locations[school_locations['LOCATION_CODE'].str.contains('M')]
```

```python
mn_school_locations.rename(columns=lambda x: x.strip())
mn_school_locations[['DBN',"lat", "lng"]].to_csv('./processed/school_locations.csv', index= False)
```
